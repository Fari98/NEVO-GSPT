```python
fig, axes = plt.subplots(2, 2, figsize=(12, 8))  # Adjust as needed
axes = axes.flatten()

for idx, dataset in enumerate(datas):
    ax = axes[idx]

    selected = data[(data['dataset'] == dataset) & 
                    (data['generation'] == 2000)]['test_fitness'].values.tolist()

    selected2 = data_after_train[data_after_train['dataset'] == dataset]['test_fitness'].values.tolist()

    # selected3 = baseline_final[baseline_final['dataset'] == dataset]['test_fitness'].values.tolist()

    ax.boxplot([selected, selected2],
               # labels=['NEVO-GSPT', 'NEVO-GSPT + backpropagation', 'MLP']
                # colors = ['red', 'blue', 'green']
              )

    colors = ['#1f77b4', '#ff7f0e']  # Custom RGB hex or names
    whis_colors = ['#1f77b4', '#1f77b4', '#ff7f0e', '#ff7f0e']
    box = ax.boxplot([selected, selected2],
                     patch_artist=True
               # labels=['NEVO-GSPT', 'NEVO-GSPT + backpropagation', 'MLP']
                # colors = ['red', 'blue', 'green']
              )

    for patch in box['boxes']:
        patch.set_facecolor('none') 
    
    for patch, color in zip(box['boxes'], colors):
        # patch.set(facecolor=color)
        patch.set_edgecolor(color)
    
    for  median, color in zip(box['medians'], colors):
        median.set(color=color, linewidth=2)

    for whisker, color in zip(box['whiskers'], whis_colors):
        whisker.set(color=color, linewidth=1.5)  # Set to match your desired color

    for cap, color in zip(box['caps'], whis_colors):
        cap.set(color=color, linewidth=1.5)

    for flier, color in zip(box['fliers'], colors):
        flier.set(marker='o',  markeredgecolor=color, markersize=6)

    ax.set_xticks([])
    ax.set_title(dataset)
    ax.set_ylabel('Test RMSE')

plt.tight_layout()
# plt.show()
print(matplot2tikz.get_tikz_code())
```
