```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
paths = list(filter(lambda x: '20250521' in x, os.listdir('../log')))
```

```python
data = pd.read_csv('../log/20250521_ms_0.5.csv', names = ['dataset', 'train_fitness', 'test_fitness', 'parameters',
                                                        'depth', 'width', 'seed', 'generation', 'timing'])
data['ms'] = '0.5'
for path in paths[1:]:
    new_df =  pd.read_csv('../log/' + path, names = ['dataset', 'train_fitness', 'test_fitness', 'parameters',
                                                        'depth', 'width', 'seed', 'generation', 'timing'])
    new_df['ms'] = path.split('_')[-1].split('.')[0]
    data = pd.concat((data, new_df), axis = 0)
data.reset_index(drop = True, inplace = True)
```

```python
baseline_final = pd.read_csv('../log/baseline_20250522.csv', names = ['dataset',  'test_fitness', 'seed','epoch', 'timing'])
```

```python
baseline_evo =  pd.read_csv('../log/baseline_evo_20250522.csv', names = ['dataset', 'train_fitness', 'val_fitness', 'test_fitness', 'seed','epoch', 'timing'])
baseline_evo.dropna(axis = 0, inplace = True)
```

```python
data.drop_duplicates(subset=['dataset', 'seed', 'generation', 'ms'], keep = 'last', inplace = True)
```

```python
datas = data.dataset.unique()
msteps = data.ms.unique()
```

```python
colors = ['#FF0000', '#0000FF', '#008000', '#800080']
light_colors = ['#F08080', '#87CEFA', '#90EE90', '#DDA0DD']
```

```python
baseline_final.head()
```

```python
for dataset in datas:
    for i, ms in enumerate(msteps):
        
        grouped = data[(data['dataset'] == dataset) &
                       (data['ms'] == ms)][
                        ['train_fitness', 'test_fitness', 'generation']].groupby('generation').median()
        
        plt.plot(grouped.index, grouped.train_fitness, label = ms, color = colors[i])
        plt.plot(grouped.index, grouped.test_fitness, color = light_colors[i], linestyle = 'dashed') # label = 'test',


    baseline = baseline_final[baseline_final.dataset == dataset][['test_fitness', 'epoch']].groupby('epoch').median().values[0]
    
    plt.plot([i for i in range(5000)], [baseline for _ in range(5000)], color = 'black', linestyle = 'dotted', label = 'NN baseline')

    plt.title(dataset)
    plt.xlabel('Generation')
    plt.ylabel('RMSE')
    plt.legend()
    plt.show()
```

```python
for dataset in datas:
        
    grouped = baseline_evo[(baseline_evo['dataset'] == dataset)][
                    ['train_fitness', 'val_fitness', 'test_fitness', 'epoch']].groupby('epoch').median()
    
    plt.plot(grouped.index, grouped.train_fitness, label = 'train')
    plt.plot(grouped.index, grouped.val_fitness, label = 'val') # label = 'test',
    plt.plot(grouped.index, grouped.test_fitness, label = 'test')


    # baseline = baseline_final[baseline_final.dataset == dataset][['test_fitness', 'epoch']].groupby('epoch').median().values[0]
    
    # plt.plot([i for i in range(5000)], [baseline for _ in range(5000)], color = 'black', linestyle = 'dotted', label = 'NN baseline')

    plt.title(dataset)
    plt.xlabel('Generation')
    plt.ylabel('MAE')
    plt.legend()
    plt.show()
```
